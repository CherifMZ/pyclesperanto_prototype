```python
from skimage.io import imread
import pyclesperanto_prototype as cle
```

When working with numpy / scikit-image, images are per default visualized as array of numbers.

```python
blobs = imread("../../data/blobs.tif")
blobs
```

In pyclesperanto_prototype, we per default visualize images as images, with additional information such as shape, pixel type and size.

```python
cle_blobs = cle.asarray(blobs)
cle_blobs
```

For customized visualization of the image, e.g. colormaps and colorbars, use `imshow()`.

```python
cle.imshow(cle_blobs, colormap='jet')
```

If the image is small (number of pixels < 100), the values are also printed instead.

```python
data = cle.asarray([[0, 1],[2, 3]])
data
```

The visualization interprets `uint32` pixel type as label image and visualizes label images in color.

```python
labels = cle.voronoi_otsu_labeling(blobs, spot_sigma=3.5)
```

```python
labels
```

You can supress that behaviour by changing the type of the image.

```python
labels.astype(float)
```

Three dimensional images are visualized as maximum projection.

```python
head = cle.asarray(imread("../../data/Haase_MRT_tfl3d1.tif"))
head
```

You can visualize indiviual slices like when slicing numpy-arrays.

```python
head[60]
```

```python
head[:, 100]
```

```python
head[:,:,100]
```
