```python
import pyclesperanto_prototype as cle
import numpy as np
from numpy import random
from skimage.io import imshow
import matplotlib
```

```python
pointlist = cle.push_zyx(random.random((2,25)) * 100) 
print(pointlist)
```

# Starting point: Label map

```python
labelled_spots = cle.pointlist_to_labelled_spots(pointlist)
cells = cle.extend_labeling_via_voronoi(labelled_spots)

cmap = matplotlib.colors.ListedColormap ( np.random.rand ( 256,3))
imshow(cle.pull_zyx(cells), cmap=cmap)
```

## Nearest neighbor distance maps

```python
average_distance_of_n_closest_neighbors_map = cle.average_distance_of_n_closest_neighbors_map(cells, n=1)
imshow(cle.pull_zyx(average_distance_of_n_closest_neighbors_map), cmap='jet')
```

```python
average_distance_of_n_closest_neighbors_map = cle.average_distance_of_n_closest_neighbors_map(cells, n=5)
imshow(cle.pull_zyx(average_distance_of_n_closest_neighbors_map), cmap='jet')
```

## Touching neighbor distance map

```python
average_neighbor_distance_map = cle.average_neighbor_distance_map(cells)
imshow(cle.pull_zyx(average_neighbor_distance_map), cmap='jet')
```
