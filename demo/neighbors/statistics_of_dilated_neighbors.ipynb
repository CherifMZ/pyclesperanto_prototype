# Neighborhood statistics of dilated label images


```python
import pyclesperanto_prototype as cle
import pandas as pd

cle.select_device("TX")
```

```python
cle.__version__
```

Assume we have a label image like this. Obviously, the labels are not touching.

```python
label_image = cle.asarray([
    [0, 0, 0, 0, 0],
    [0, 1, 0, 2, 0],
    [0, 0, 0, 0, 0],
])
```

```python
stats = cle.statistics_of_labelled_neighbors(label_image, 
                                             proximal_distances=[], 
                                             nearest_neighbor_ns=[], 
                                             dilation_radii=[0, 1, 2])

df = pd.DataFrame(stats)
```

```python
df.describe().T
```

The resulting table contains columns for neighbor statistics of dilated labels with specified radii (1,2). From these columns we can see how these statistics change when labels are dilated.

```python
df[["label", "touch_count_sum_dilated_r_0", "touch_count_sum_dilated_r_1", "touch_count_sum_dilated_r_2"]]
```

```python
df[["label", "average_distance_of_touching_neighbors_dilated_r_0", "average_distance_of_touching_neighbors_dilated_r_1"]]
```

Visually check if these numbers make sense:

```python
print(cle.dilate_labels(label_image, radius=1))
```
