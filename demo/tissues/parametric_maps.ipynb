```python
import pyclesperanto_prototype as cle

from skimage.io import imread, imsave, imshow
import matplotlib
import numpy as np
```

```python
# load data
image = imread('https://samples.fiji.sc/blobs.png')
blobs = cle.push_zyx(image)

imshow(cle.pull_zyx(blobs))
```

```python
binary = cle.threshold_otsu(blobs)

imshow(cle.pull_zyx(binary))
```

```python
labels = cle.connected_components_labeling_box(binary)

cmap = matplotlib.colors.ListedColormap ( np.random.rand ( 256,3))
imshow(cle.pull_zyx(labels),  cmap=cmap)
```

# Quantitative maps a.k.a. parametric images
## Pixel count map

```python
pixel_count_map = cle.label_pixel_count_map(labels)
imshow(cle.pull_zyx(pixel_count_map), cmap='jet')
```

# Mean/min/max intensity map

```python
mean_intensity_map = cle.label_mean_intensity_map(blobs, labels)
imshow(cle.pull_zyx(mean_intensity_map), cmap='jet')
```

```python
minimum_intensity_map = cle.label_minimum_intensity_map(blobs, labels)
imshow(cle.pull_zyx(minimum_intensity_map), cmap='jet')
```

```python
maximum_intensity_map = cle.label_maximum_intensity_map(blobs, labels)
imshow(cle.pull_zyx(maximum_intensity_map), cmap='jet')
```
