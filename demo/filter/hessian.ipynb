```python
import pyclesperanto_prototype as cle

cle.select_device("RTX")
```

```python
image = cle.artificial_objects_2d()
cle.imshow(image)
```

```python
blurred = cle.gaussian_blur(image, sigma_x=1, sigma_y=1)
cle.imshow(blurred)
```

```python
h1, h2, h3 = cle.hessian_eigenvalues(blurred)
cle.imshow(h1, colorbar=True)
cle.imshow(h3, colorbar=True)
```

```python
h1,h2,h3 = cle.hessian_eigenvalues([image, image, image])
cle.imshow(h1, colorbar=True)
cle.imshow(h2, colorbar=True)
cle.imshow(h3, colorbar=True)
```
